```python
# Reading an Image and display using open cv

import cv2
import os

def read_and_display_opencv(image_path):
    """
    Reads and displays an image using OpenCV
    """
    img = cv2.imread(image_path)

    if img is None:
        raise ValueError("Image not found or path is incorrect")

    cv2.imshow("OpenCV Image", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# Example usage
base_dir  = os.getcwd()
print(base_dir)
read_and_display_opencv("Macho.jpg")
```

```python
from PIL import Image

def read_and_display_pil(image_path):
    """
    Reads and displays an image using PIL
    """
    img = Image.open(image_path)

    img.show()


# Example usage
# read_and_display_pil("sample.jpg")
```

```python
import cv2
from PIL import Image

def opencv_read_pil_display(image_path):
    """
    Reads image with OpenCV and displays using PIL
    """
    img_bgr = cv2.imread(image_path)

    if img_bgr is None:
        raise ValueError("Image not found")

    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(img_rgb)

    img_pil.show()


# Example usage
# opencv_read_pil_display("sample.jpg")
```
